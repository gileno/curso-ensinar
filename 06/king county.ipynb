```python
X = dados[
    ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors', 'waterfront', 'view', 'yr_built',
     'grade', 'condition']
]
```

```python
import statsmodels.api as sms
```

```python
Y = dados['price']
modelo = sms.OLS(Y, X)
resultado = modelo.fit()
```

```python
resultado.summary()
```

```python
df = pd.DataFrame(columns=['Y', 'Yhat'])
```

```python
predicoes = resultado.predict(X)
```

```python
df['Yhat'] = predicoes
df['Y'] = Y
```
