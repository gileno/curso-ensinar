```python
media_iptu = X[X['iptu'] != 0]['iptu'].mean()
media_area = X[X['iptu'] != 0]['area'].mean()
```

```python
iptu_m2 = media_iptu / media_area
def preencher_iptu(linha):
    if linha['iptu'] == 0:
        return iptu_m2 * linha['area']
    return linha['iptu']
```

```python
X['iptu'] = X.apply(preencher_iptu, axis=1)
```

```python
X.head()
```

```python
modelo = sm.OLS(Y, X)
resultado = modelo.fit()
resultado.summary()
```
