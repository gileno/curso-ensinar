```python
def limpar_valores(valor):
    resultado = ""
    for letra in valor:
        if letra.isdigit():
            resultado = resultado + letra
    if resultado:
        return int(resultado)
    return 0
```

```python
dados['iptu'] = dados['iptu'].fillna("0")
dados['condominio'] = dados['condominio'].fillna("0")
dados['suites'] = dados['suites'].fillna("0")
```

```python
import folium
from IPython.display import display
centro = (dados.iloc[0]['latitude'], dados.iloc[0]['longitude'])
mapa = folium.Map(location=centro, zoom_start=13)
dados = dados.fillna(0)
for i in range(len(dados)):
    linha = dados.iloc[i]
    if linha['latitude'] == 0:
        continue
    ponto = (linha['latitude'], linha['longitude'])
    marcador = folium.Marker(ponto, popup=linha['link'])
    marcador.add_to(mapa)
display(mapa)
```
