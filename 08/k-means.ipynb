```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X, Y = make_blobs(
    n_samples=300, centers=4, cluster_std=0.60
)
```

```python
X[:10]
```

```python
Y
```

```python
dados = pd.DataFrame(X, columns=['x', 'y'])
dados['cluster'] = Y
```

```python
dados.head()
```

```python
plt.scatter(dados['x'], dados['y'])
```

```python
plt.scatter(dados['x'], dados['y'], c=Y)
```

```python
from sklearn.cluster import KMeans
```

```python
modelo = KMeans(n_clusters=2)
modelo.fit(dados[['x', 'y']])
```

```python
modelo.labels_
```

```python
plt.scatter(dados['x'], dados['y'], c=modelo.labels_)
```

```python
modelo.inertia_
```

```python
inertia = []
for i in range(2, 8):
    modelo = KMeans(n_clusters=i)
    modelo.fit(dados[['x', 'y']])
    inertia.append(modelo.inertia_)
plt.plot(range(2, 8), inertia)
```
